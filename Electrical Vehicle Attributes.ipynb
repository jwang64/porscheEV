## How does Porsche compare to the top performers in each attribute?

```python
quantitative_attributes = ['charge240','cityE','rangeA','rangeCity','rangeCityA','rangeHwy','rangeHwyA','charge240b']
```

```python
for attribute in quantitative_attributes:
    y = ev_attributes[attribute]
    string = ev_attributes['Make'] + " " + ev_attributes['Model']
    x = string
    plt.figure(figsize=(30,10))
    title_string = attribute + " for each car"
    plt.title(title_string)
    bar_chart = plt.bar(x,y)
    bar_chart[10].set_color('r')
    bar_chart[11].set_color('r')
    plt.ylabel(attribute)
    plt.xlabel("Make and Model")
    plt.tick_params(axis='x',which='major', labelsize= 8)
    plt.show()
```
